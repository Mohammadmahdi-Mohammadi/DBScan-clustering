### Part B & C

```python
def DBScan(X_train: np.ndarray, Epsilon, MinPoints):
    m, _ = X_train.shape
    result = np.zeros(m, dtype=int)
    C = 0
    for i in tqdm(range(m)):
        if result[i] != 0:
            continue
        m, n = X_train.shape
        neighbors = []
        for j in range(m):
            if i == j:
                continue
            if math.sqrt(np.sum(np.power(X_train[i, :] - X_train[j, :], 2))) < Epsilon:
                neighbors.append(j)
        if len(neighbors) < MinPoints:
            result[i] = -1  
            continue
        C += 1
        result[i] = C
        while len(neighbors) != 0:
            index_array = neighbors.pop(0)
            if result[index_array] == -1:
                result[index_array] = C  
            if result[index_array] != 0:
                continue
            result[index_array] = C
            m, _ = X_train.shape
            new_neighbors = []
            for j in range(m):
                if index_array == j:
                    continue
                if math.sqrt(np.sum(np.power(X_train[index_array, :] - X_train[j, :], 2))) < Epsilon:
                    new_neighbors.append(j)
            if len(new_neighbors) >= MinPoints: 
                neighbors.extend(new_neighbors)
    return result
```

```python
directory = "input/P3/D1.txt"
main,final_target = [],[]
with open(directory) as drct:
    for item in drct.readlines():
        if item.__contains__("%") or item.__contains__("@") or item.__eq__("\n"):
            continue
        main.append([float(item[:-1].split(",")[0]), float(item[:-1].split(",")[1])])
        if float(item[:-1].split(",")[2] == "noise"):
            final_target.append(-1)
        else:
            final_target.append(int(float(item[:-1].split(",")[2])))
data = np.array(main)
final_target = np.array(final_target)
current_labels = DBScan(np.array(main), 4, 5)
print("num of clusters: ", max(current_labels))
colors = []
for i in range(200):
    colors.append('#%06X' % randint(0, 0xFFFFFF))
plt.scatter(data[:, 0], data[:, 1], c=list(map(lambda x: colors[x], current_labels)), marker="o", picker=True)
plt.show()
```
